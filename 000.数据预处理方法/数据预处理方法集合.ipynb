#  插值方法： ###
- ffill 空值取前面的值
- bfill 空值取后面的值
- interpolate 线性取值

```python
df=df.resample('D').ffill(1)
df=df.resample('D').bfill(1)
```

```python
df=df.resample('D').interpolate('linear')
```
